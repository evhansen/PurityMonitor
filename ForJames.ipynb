We need to get the data into the program. If we were talking tens or even hundreds of data points, you could do this simply in python. However since each file is a million lines long, we're going to use the "pandas" package which handles big datasets better. 


```python
#label the columns that you're importing (we will use skiprows later to ignore labels from the CSV file itself)
column_names = ['time','V']

#read in the CSV file
dataA = pd.read_csv('May4_Rewire/AnodeOnlyVLite.CSV', skiprows=1, names=column_names)

#the following line converts the time value from seconds to microseconds so that we don't run into
#any underflow errors. Remember this when you're labeling your plot!
dataA.time*=1E6

#print out the dataset so we know it worked
dataA
```

```python
#make a line plot of the data. There are many ways to plot 'pandas' dataframes. Look up the documentation for 
#'pandas' and 'seaborn' for all of your options.
dataA.plot('time','V')
plt.xlim([0,20])
#plt.ylim([-0.04,0])
```
