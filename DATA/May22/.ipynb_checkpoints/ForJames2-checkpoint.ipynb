```python
%matplotlib inline 

import math
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd 
import matplotlib.cm as cm

plt.rcParams['figure.figsize'] = [16, 10]
plt.rcParams.update({'font.size': 18})
current_palette = sns.color_palette()
```

```python
column_names = ['time','Cathode', 'Anode']

dataA = pd.read_csv('field_data_0.txt', sep = '\t',skiprows=1, names=column_names)
dataB = pd.read_csv('field_data_1.txt', sep = '\t', skiprows=1, names=column_names)
dataC = pd.read_csv('field_data_2.txt', sep = '\t', skiprows=1, names=column_names)
dataD = pd.read_csv('field_data_3.txt', sep = '\t', skiprows=1, names=column_names)
dataE = pd.read_csv('field_data_4.txt', sep = '\t', skiprows=1, names=column_names)
dataF = pd.read_csv('field_data_5.txt', sep = '\t', skiprows=1, names=column_names)

dataA.time*=1E6
dataB.time*=1E6
dataC.time*=1E6
dataD.time*=1E6
dataE.time*=1E6
dataF.time*=1E6
```

```python
dataA1 = dataA.copy(deep=False)
dataA2 = dataA.copy(deep=False)
dataA3 = dataA.copy(deep=False)
del dataA1['Anode']
del dataA2['Cathode']
dataA3['Sum']=dataA3['Cathode'] + dataA3['Anode']
del dataA3['Cathode']
del dataA3['Anode']
ax1=dataA1.plot('time', 'Cathode', title="Field Data 0", label="Cathode")
dataA2.plot('time', 'Anode', label="Anode",ax=ax1)
dataA3.plot('time', 'Sum', label="Sum",ax=ax1)
plt.xlim([0,30])
plt.ylim([-.2,.2])

dataB1 = dataB.copy(deep=False)
dataB2 = dataB.copy(deep=False)
dataB3 = dataB.copy(deep=False)
del dataB1['Anode']
del dataB2['Cathode']
dataB3['Sum']=dataB3['Cathode'] + dataB3['Anode']
del dataB3['Cathode']
del dataB3['Anode']
ax2=dataB1.plot('time', 'Cathode', title="Field Data 1", label="Cathode")
dataB2.plot('time', 'Anode', label="Anode",ax=ax2)
dataB3.plot('time', 'Sum', label="Sum",ax=ax2)
plt.xlim([0,30])
plt.ylim([-.2,.2])

dataC1 = dataC.copy(deep=False)
dataC2 = dataC.copy(deep=False)
dataC3 = dataC.copy(deep=False)
del dataC1['Anode']
del dataC2['Cathode']
dataC3['Sum']=dataC3['Cathode'] + dataC3['Anode']
del dataC3['Cathode']
del dataC3['Anode']
ax3=dataC1.plot('time', 'Cathode', title="Field Data 2", label="Cathode")
dataC2.plot('time', 'Anode', label="Anode",ax=ax3)
dataC3.plot('time', 'Sum', label="Sum",ax=ax3)
plt.xlim([0,30])
plt.ylim([-.2,.2])

dataD1 = dataD.copy(deep=False)
dataD2 = dataD.copy(deep=False)
dataD3 = dataD.copy(deep=False)
del dataD1['Anode']
del dataD2['Cathode']
dataD3['Sum']=dataD3['Cathode'] + dataD3['Anode']
del dataD3['Cathode']
del dataD3['Anode']
ax4=dataD1.plot('time', 'Cathode', title="Field Data 3", label="Cathode")
dataD2.plot('time', 'Anode', label="Anode",ax=ax4)
dataD3.plot('time', 'Sum', label="Sum",ax=ax4)
plt.xlim([0,30])
plt.ylim([-.2,.2])

dataE1 = dataE.copy(deep=False)
dataE2 = dataE.copy(deep=False)
dataE3 = dataE.copy(deep=False)
del dataE1['Anode']
del dataE2['Cathode']
dataE3['Sum']=dataE3['Cathode'] + dataE3['Anode']
del dataE3['Cathode']
del dataE3['Anode']
ax5=dataE1.plot('time', 'Cathode', title="Field Data 4", label="Cathode")
dataE2.plot('time', 'Anode', label="Anode",ax=ax5)
dataE3.plot('time', 'Sum', label="Sum",ax=ax5)
plt.xlim([0,30])
plt.ylim([-.2,.2])

dataF1 = dataF.copy(deep=False)
dataF2 = dataF.copy(deep=False)
dataF3 = dataF.copy(deep=False)
del dataF1['Anode']
del dataF2['Cathode']
dataF3['Sum']=dataF3['Cathode'] + dataF3['Anode']
del dataF3['Cathode']
del dataF3['Anode']
ax6=dataF1.plot('time', 'Cathode', title="Field Data 5", label="Cathode")
dataF2.plot('time', 'Anode', label="Anode",ax=ax6)
dataF3.plot('time', 'Sum', label="Sum",ax=ax6)
plt.xlim([0,30])
plt.ylim([-.2,.2])
```
